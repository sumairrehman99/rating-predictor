```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('updated_movies.csv')
```

```python
df_feats = df.iloc[:1000]
```

```python
def director(name):
    if name not in top_10_directors:
        name = 'Other'
    return name
```

```python
df_feats['director'] = df_feats['director'].apply(director)
```

```python
df_feats = pd.get_dummies(df_feats, columns=['director'])
```

```python
df_feats.head()
```

```python
df_feats.drop(['title', 'genre', 'stars', 'certificate', 'votes', 'gross'], inplace=True, axis=1)
df_feats.head()
```

```python
df_feats.columns
```

```python
X = df_feats[['year', 'runtime', 'metascore',
       'director_Alan Parker', 'director_Alejandro G. Iñárritu',
       'director_Alfonso Cuarón', 'director_Alfred Hitchcock',
       'director_Ang Lee', 'director_Anthony Russo', 'director_Brad Bird',
       'director_Christopher Nolan', 'director_Clint Eastwood',
       'director_Danny Boyle', 'director_David Fincher',
       'director_David Lynch', 'director_David Yates',
       'director_Denis Villeneuve', 'director_Edgar Wright',
       'director_Edward Zwick', 'director_Francis Ford Coppola',
       'director_Fred Zinnemann', 'director_Guy Ritchie',
       'director_Hayao Miyazaki', 'director_Hirokazu Koreeda',
       'director_James Cameron', 'director_James Mangold',
       'director_Joel Coen', 'director_John McTiernan',
       'director_Krzysztof Kieslowski', 'director_Lars von Trier',
       'director_Marc Forster', 'director_Martin Scorsese',
       'director_Michael Curtiz', 'director_Other', 'director_Paul Greengrass',
       'director_Pedro Almodóvar', 'director_Peter Jackson',
       'director_Quentin Tarantino', 'director_Richard Linklater',
       'director_Ridley Scott', 'director_Rob Reiner',
       'director_Robert Altman', 'director_Robert Zemeckis',
       'director_Ron Howard', 'director_Satoshi Kon', 'director_Sergio Leone',
       'director_Sidney Lumet', 'director_Spike Lee',
       'director_Stanley Kubrick', 'director_Steven Spielberg',
       'director_Terry Gilliam', 'director_Tim Burton',
       'director_Wes Anderson', 'director_Woody Allen']].values
y= df_feats['rating'].values
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(54, activation='relu'))
model.add(Dense(54, activation='relu'))
model.add(Dense(54, activation='relu'))
model.add(Dense(54, activation='relu'))

model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=128, epochs=600)
```

```python
history = pd.DataFrame(model.history.history)
history.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
explained_variance_score(y_test, predictions)
```
